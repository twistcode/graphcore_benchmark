```python
plt.rcParams["figure.figsize"] = [7.50, 3.50]
plt.rcParams["figure.autolayout"] = True
fig = plt.figure()
```

```python
pemecut = pd.DataFrame({'Nvidia GV100': [2, 31.5],
                         'Nvidia Titan V': [2, 33.5],
                         'Intel Xeon Gold 6152': [4, None],
                        'Graphcore IPU': [1.26, 17]},
                        index=['MNIST', 'RSNA'])
pemecut.plot(kind="bar")
# ax = plt.axes()
# ax.set_facecolor("yellow")
plt.title("Tensorflow 2  Classification Training Runtime")
plt.ylabel("Minutes")
resolution_value = 1200
plt.savefig('fig_1.jpg', format="jpg", dpi=resolution_value)
```

```python
pemecut2 = pd.DataFrame({'Nvidia GV100': [51, 81],
                         'Nvidia Titan V': [58, 84],
                         'Intel Xeon Gold 6152': [None, None],
                        'Graphcore IPU': [29, 82]},
                        index=['MNIST', 'RSNA'])
bar_plot=pemecut2.plot(kind="bar")
bar_plot.axhline(y=60, color= 'red', linewidth=2 , linestyle='dashed')
# add text
bar_plot.text(x= .2, y=61.25, s='IPU Safe Limit')
plt.title("Tensorflow 2 Classification Training")
plt.ylabel("Celcius")
resolution_value = 1200
plt.savefig('fig_2.jpg', format="jpg", dpi=resolution_value)
```
